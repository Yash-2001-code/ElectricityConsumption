```python
train_size = int(len(f1_19_daywise)*0.8)
test_size = len(f1_19_daywise)-train_size
```

```python
train_set = f1_19_daywise[:train_size]
test_set = f1_19_daywise[train_size:]
print(len(train_set))
print(len(test_set))
train_set = np.array(train_set)
test_set = np.array(test_set)

# train_size = int(len(dataset) * 0.67)
# test_size = len(dataset) - train_size
# train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
# print(len(train), len(test))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(train_set[:,-1].reshape(-1,1))
scaled_train = scaler.transform(train_set[:,-1].reshape(-1,1))
scaled_test = scaler.transform(test_set[:,-1].reshape(-1,1))
```

```python
scaled_test
```

```python
print(scaled_train.shape)
print(scaled_test.shape)
```

```python
def create_dataset(dataset, time_step = 1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step),0]
        dataX.append(a)
        dataY.append(dataset[i+time_step,0])
    return np.array(dataX), np.array(dataY)
```

```python
time_step = 28
X_train, y_train = create_dataset(scaled_train, time_step)
X_test, y_test=  create_dataset(scaled_test, time_step)
```

```python
print(X_train)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1],1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1],1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
n_input = 28

generator = TimeseriesGenerator(X_train, X_train, length=n_input, batch_size=1)
```

```python
print('Samples: %d' % len(generator))
```

```python
x,y = generator[0]
print(x.flatten())
print(y)
```

```python
n_features = 1
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features),return_sequences=True))
model.add(LSTM(100, activation='relu'))
model.add(Dense(25))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train,y_train, validation_data = (X_test, y_test), epochs=50, batch_size=64, verbose=1)
```

```python
model.summary()
```
